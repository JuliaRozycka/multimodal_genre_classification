```python
# Check folder spectrograms data, for each foldern (genres) in spectrograms data, count number of files and insert into a dictionry [genre: no files]

import os
import numpy as np

path = 'spectograms_data'
genres = os.listdir(path)
genres_dict = {}
for genre in genres:
    genres_dict[genre] = len(os.listdir(os.path.join(path, genre)))
genres_dict
```

```python
# Plot the genre distribution
import matplotlib.pyplot as plt
import random
import matplotlib as mpl

cmap = mpl.colormaps['viridis']
colors = cmap(np.linspace(0, 1, 20)).tolist() 

# Plot the colors
plt.figure(figsize=(8, 2))
for i,  color in enumerate(colors):
    plt.barh(i, width = 0.5, color=color)
    #plt.text(0.5, i, color, va='center', ha='center', color='white', fontsize=12)

plt.title("Genre Colors from Viridis Colormap")
plt.show()


```

```python
# Plot the genre distribution
import matplotlib.pyplot as plt
import random
import matplotlib as mpl

cmap = mpl.colormaps['viridis']
colors = cmap(np.linspace(0, 1, len(genres))).tolist() # Generate colors from colormap


plt.figure(figsize=(10, 5))
plt.bar(genres_dict.keys(), genres_dict.values(), color=colors[1])
plt.xlabel('Genres')
plt.ylabel('Number of files')

# Use tex
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

plt.show()
```
